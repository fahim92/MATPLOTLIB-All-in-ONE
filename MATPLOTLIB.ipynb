Drawing a plot using numpy array

```python
import numpy as np
import matplotlib.pyplot as plt
```

Let's create an array using arange function

```python
x=np.arange(1,11)
x
```

```python
y=3*x
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.title('line graph')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot(x,y,color='g',linestyle=':',linewidth=2)
plt.xlabel('This is x axis')
plt.ylabel('This is y axis')
plt.title('Line graph')
plt.show()
```

# Drawring 2 lines in one graph


```python
x=np.arange(1,12)
y1=2*x
y2=3*x
plt.plot(x,y1,color='g',linestyle=':',linewidth=5)
plt.plot(x,y2,color='y',linestyle='-',linewidth=5)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid(True)
plt.show()
```

# Drawing subplots
Drawing subplot means drawing multiple plots side by side

```python
x=np.arange(1,12)
y1=2*x
y2=3*x
plt.subplot(1,2,1)
plt.plot(x,y1,color='g',linestyle=':',linewidth=5)
plt.subplot(1,2,2)
plt.plot(x,y2,color='y',linestyle='-',linewidth=5)
plt.grid(True)
plt.show()
```

# Bar plot
Bar plot is usually used for categorical distribution. It can be created with the help of a dictionary. It is convenient to draw bar plot from lists with the help of .keys and .values

```python
student={'Bob':87,'Julia':45,'Matt':67}
name=list(student.keys())
score=list(student.values())
```

```python
name
```

```python
score
```

```python
plt.bar(name,score,color='g')
plt.xlabel('Names of students')
plt.ylabel('Scores of students')
plt.title('Distribution of student marks')
plt.show()
```

# Horizontal Bar plot
Drawing a horizontal bar plot is almost similar to drawing a bar plot. Only difference is we need to write plt.barh instead of plt.bar. That's it

```python
plt.barh(name,score,color='g')
plt.xlabel('Names of students')
plt.ylabel('Scores of students')
plt.grid(True)
plt.title('Distribution of student marks')
plt.show()
```

# Basic Scatter Plot


```python
import numpy as np
x=[2,3,4,1,9,6,8]
y=[1,5,8,4,9,6,4]
plt.scatter(x,y,marker='*',c='g',s=100)
plt.show()
```

Here marker is used to specify stars. c stands for color and s is used for initializing size 

# Adding two markers in the same plot

```python
import numpy as np
x=[2,3,4,1,9,6,8]
a=[1,5,8,4,9,6,4]
b=[2,8,7,5,9,6,8]
plt.scatter(x,a,marker='*',c='g',s=100)
plt.scatter(x,b,marker='.',c='r',s=100)
plt.show()
```

# Scatter Subplot

```python
import numpy as np
x=[2,3,4,1,9,6,8]
a=[1,5,8,4,9,6,4]
b=[2,8,7,5,9,6,8]
plt.subplot(1,2,1)
plt.scatter(x,a,marker='*',c='g',s=100)
plt.subplot(1,2,2)
plt.scatter(x,b,marker='.',color='r',s=100)
plt.show()
```

# Histogram
There is slight differnt between bar chart and histogram. Bar chart is mainly used for categorical data. On the contrary histogram can be used for numerical values.histogram can be created from a list of random values. In a histogram you get the frequency from y axis.

```python
data=[1,3,3,7,9,4,9,5,8,8]
plt.hist(data,color='g',bins=10)
```

# Boxplot
Boxplot is useful in visualizing median values.Input values of the boxplot are taken from lists.

```python
one=[1,2,3,4,5,6,7,8,9]
two=[2,6,1,7,9,10,12,5]
three=[10,12,14,16,20]
data=list([one,two,three])
plt.boxplot(data)
plt.show()
```

# As we can see first a list of list is being generated then the values is stored in data. Finally it is fed into plt.boxplot(data) 

# Violin plot
It is almost similar to boxplot.Example is shown below

```python
one=[1,2,3,4,5,6,7,8,9]
two=[2,6,1,7,9,10,12,5]
three=[10,12,14,16,20]
data=list([one,two,three])
plt.violinplot(data)
plt.show()
```

# Pie Chart
It is an important chart for visualizing

```python
fruits=['apple','orange','mango','guava']
quantity=[90,120,40,32]
plt.pie(quantity,labels=fruits)
plt.show()
```

If we want to add percentage we can add it by typing autopct=%0.1%%

```python
fruits=['apple','orange','mango','guava']
quantity=[90,120,40,32]
plt.pie(quantity,labels=fruits)
plt.show()
```

```python
fruits=['apple','orange','mango','guava']
quantity=[90,120,40,32]
plt.pie(quantity,labels=fruits,autopct='%0.1f%%')
plt.show()
```
